```python
import jsoncfg
from jsoncfg.config_classes import ConfigJSONObject, ConfigJSONArray, ConfigJSONScalar

def recursivePrint(element):
    if isinstance(element, ConfigJSONObject):
        # Dictionary
        for key, value in element:
            print(f"key   \"{key}\" at line {jsoncfg.node_location(value).line}")
            recursivePrint(value)
    elif isinstance(element, ConfigJSONArray):
        # Array
        for item in element:
            recursivePrint(item)
    elif isinstance(element, ConfigJSONScalar):
        value = element()
        if isinstance(value, str):
            value = value.strip()

        print(f"value \"{value}\" at line {jsoncfg.node_location(element).line}")

parsed = jsoncfg.load_config(file_link)

# json_object_lines = recursivePrint(parsed)
```

```python
def print_line_numbers(element):
    if isinstance(element, dict):
        for key, value in element.items():
            # print(f"Key: '{key}' is on line {jsoncfg.node_location(element[key]).line}")
            print(f"Key: '{key}'")
            print(f"element[key] '{element[key]}'")
            print_line_numbers(value)
    elif isinstance(element, list):
        for item in element:
            print_line_numbers(item)
    elif isinstance(element, str):
        print(f"Value: '{element}' is on line {jsoncfg.node_location(element).line}")

print_line_numbers(parsed_data)
```

```python
import hashlib

message = "Hello, world!"
encoded_message = message.encode('utf-8')

# Create a SHA256 hash object
hasher = hashlib.sha256()

# Update the hash object with the message
hasher.update(encoded_message)

# Get the hexadecimal digest
hex_digest = hasher.hexdigest()

print(f"SHA256 hash of '{message}': {hex_digest}")
```

```python
from collections import Counter

# Counting elements in a list
data = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
fruit_counts = Counter(data)
print(f"Fruit counts: {fruit_counts}")

# Accessing counts
print(f"Count of 'apple': {fruit_counts['apple']}")
print(f"Count of 'grape': {fruit_counts['grape']}") # Returns 0 for missing key

# Finding most common elements
most_common_fruits = fruit_counts.most_common(2)
print(f"Two most common fruits: {most_common_fruits}")
```
